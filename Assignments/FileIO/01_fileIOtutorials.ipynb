```python
# Create a textfile and write text to it
with open('spandaai_training_files.txt', 'w') as file:
    file.write('Line 1: Houston - Texas\n'
               'Line 2: Austin - Texas\n'
               'Line 3: Dallas - Texas\n'
               'Line 4: San Antonio - Texas\n')

# with command simplifies the resource mgmt.
# it automatically opens and close the file
# no need of file - open(...), file.close() operations
# in case of exceptions, it closes the file automatically 
```

```python
# same code without using with
file = open('spandaai_training_files.txt', 'w')
try:
    file.write('Line 1: Houston - Texas\n'
               'Line 2: Austin - Texas\n'
               'Line 3: Dallas - Texas\n'
               'Line 4: San Antonio - Texas\n')
finally:
    file.close()  # You must manually close the file
```

```python
# Append a file
with open('spandaai_training_files.txt', 'a') as file:
    file.write('Line 5: Temple - Texas\n')
```

```python
# Reading Lines into a python List
with open('spandaai_training_files.txt', 'r') as file:
    lines = file.readlines()
    print(lines)
```

```python
# Iterating Over Each Line in a File for line level processing
with open('spandaai_training_files.txt', 'r') as file:
    for line in file:
        print(line.strip())
# Observe the difference between List output vs Iterative reads
```

```python
# check if a file exists before performing file operations:
import os
if os.path.exists('spandaai_training_files.txt'):
    print('File exists.')
else:
    print('File does not exist.')
```

```python
# Write each element of a list from a process to a new line in a file:
# Lines is a list
# Open the file in write mode
lines = ['From a List - line 1', 'From a List - line 2', 'From a List - line 3']
with open('spandaai_training_file_lists.txt', 'w') as file:
    for line in lines:
        file.write(f'{line}\n')
```

```python
# work with multiple files simultaneously using with blocks:
# open the source file with in mode and 
with open('source.txt', 'r') as source, open('destination.txt', 'w') as destination:
    content = source.read()
    destination.write(content)
```

```python
# safely delete a file if it exists:
import os
if os.path.exists('source.txt'):
    os.remove('source.txt')
    print('File deleted.')
else:
    print('File does not exist.')
```

```python
# read from and write to a file in binary mode (useful for images, videos, etc.):
# Reading a binary file
with open('Sitaram_Driving_licence.jpg', 'rb') as file:
    content = file.read()
# Writing to a binary file
with open('Sitaram_Driving_licence_copy.jpg', 'wb') as file:
    file.write(content)
```
